```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
education_districtwise = pd.read_csv("../datasets/education_districtwise.csv").dropna()
education_districtwise.head()
```

```python
sampled_data = education_districtwise.sample(n=50, replace=True)
sampled_data
```

```python
sample_mean = sampled_data["OVERALL_LI"].mean()
sample_std = sampled_data["OVERALL_LI"].std()
sample_std_err = sample_std / np.sqrt(len(sampled_data))
```

```python
#calculating the confidence interval with confidence level 95%
conf_interval = stats.norm.interval(confidence=0.95, loc=sample_mean, scale=sample_std_err)
conf_interval
```

```python
#calculating the confidence interval with confidence level 99%
conf_interval = stats.norm.interval(confidence=0.99, loc=sample_mean, scale=sample_std_err)
conf_interval
```
